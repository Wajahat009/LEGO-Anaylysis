```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
```

```python
theme_df=pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
merged_df=df.merge(theme_df , left_on='parent_theme'  , right_on='name')
```

```python
merged_df.drop(columns=['parent_theme'] , inplace=True)
```

```python
merged_df.rename(columns={'name_y' : 'parent_theme'} , inplace=True)
```

```python
merged_df.to_csv('Lego Dataset.csv')
```
